Unfortunately, the graphs you have made in the previous sections, while they may allow us to see relationships between data to a certain extent, they are not as useful as scientists need them to be. When astomoners are graphing star data they use a specific kind of graph called a Hertzsprung-Russel Diagram.
![https://i.ytimg.com/vi/Yfbs0xydBWE/maxresdefault.jpg](https://i.ytimg.com/vi/Yfbs0xydBWE/maxresdefault.jpg)

A Hertzsprung- Russel diagram is a scatterplot if stars showing the relationship between luminosity or absolute magnitude vs stellar classification or effective tempurature. These types of graphs help astronomers understand the steps of stellar evolution. In the data table above, you have everything you need to create your own Hertzsprung-Russel Diagram. Use the block of code below and the image above as a guide to create your graph.

```
plt.scatter(stars[""], stars[""], s = 1, alpha = .5, edgecolors = "none")
plt.xlabel('Something Goes Here')
plt.ylabel('Something Goes Here Too')
plt.title('You Guessed it.. Something here too')
plt.show()
```
