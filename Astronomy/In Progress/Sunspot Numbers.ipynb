# **Sunspot Number by Date**

# Data description:
The Wolf relative sunspot number derived by the formula:

**R = N<sub>s</sub> + 10 * N<sub>g</sub>,**

with N<sub>s</sub> the number of spots and N<sub>g</sub> the number of groups counted over the entire solar disk. The data used is from the Sunspot Index and Long-Term Solar Observations website. https://www.sidc.be/silso/home


# Daily Analysis
No daily data are provided before 1818 because daily observations become too sparse in earlier years. Therefore, Wolf only compiled monthly means and yearly means for all years before 1818.

The missing values are marked by -1 (valid Sunspot Number are always positive).

Filename: SN_d_tot_V2.0.csv
Format: Comma Separated values (adapted for import in spreadsheets)

Contents:
- Column 1-3: Gregorian calendar date
-- Year
-- Month
-- Day
- Column 4: Date in fraction of year.
- Column 5: Daily total sunspot number. A value of -1 indicates that no number is available for that day (missing value).
- Column 6: Daily standard deviation of the input sunspot numbers from individual stations.
- Column 7: Number of observations used to compute the daily value.
- Column 8: Definitive/provisional indicator. '1' indicates that the value is definitive. '0' indicates that the value is still provisional.

```python
# Import Libraries needed
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
#Import the total daily sunspot number data from SISLO website
spotsraw = pd.read_csv('https://github.com/rysullivan54/QuarkNet/raw/main/SN_d_tot_V20.csv')
spotsraw.head()
```

```python
#Remove all records missing data (Wolf number of -1)
spots = spotsraw.query('Daily != -1')
spots.head()
```

```python
# Create a scatter plot of all the daily sunspot data
fig, ax = plt.subplots(figsize=(90,10))
x=spots['DecDate']
y=spots['Daily']
plt.scatter(x,y, marker = '.', s= 50, color = 'black')
plt.title("Sunspot Numbers by date")
plt.ylabel("Daily Number")
plt.xlabel("Date");
```
