```
# imports some software packages we'll use
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```
#get atmospheric data and display the first few rows
data = pd.read_csv('https://scrippsco2.ucsd.edu/assets/data/atmospheric/stations/in_situ_co2/monthly/monthly_in_situ_co2_mlo.csv', skiprows=54)
data.head(5)
```

```
#remove rows 0 and 1
data = pd.DataFrame.drop(data, index=0)
data = pd.DataFrame.drop(data, index=1)
data.head(3)
```

```
#How much data do you have? Is this a sufficient amount of data?
data.shape
```

```
#Rename columns
data.columns=['Year','Month','Date','Date2','CO2','seasonally adjusted','fit','seasonal adjusted fit','CO2 filled','seasonally adjusted filled']
data.head(4)
```

```
#Observe CO2 levels over time by graphing the data

plt.scatter(data['Year'][2:30], data['CO2'][2:30]) #I used some of Enrique's code to look at just a few years at a time.
plt.title('CO2 vs. time (mos)')
plt.xlabel('Time (mos)')
plt.ylabel('CO2 ppm')
#plt.xticks([0, 250, 500, 750]) #I was trying to adjust the scale on the x and y axis to accomodate more data, but I never got there.


```
