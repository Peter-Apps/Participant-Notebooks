```python
## TEACHER'S EXAMPLE TO COMPLETE
yi = 1      #initial vertical position
vyi = 5     #initial vertical velocity
m = 0.085   #initial vertical position
g = 9.81  #vertical acceleration

# use a model for the motion of a ball tossed into the air to predict the ball's position and velocity
# calculate y positions: y = yi + vyi*t + 1/2 * a * t^2
Model['y'] = yi + vyi * Model['t'] + 0.5 * (-g) * Model['t']**2

# calculate y velocities: vy = vyi + a * t
Model['vy'] = vyi + g * Model['t']

#Model.head(3)
```

```python
# CALCULATE ENERGIES
```

```python
## TEACHER'S EXAMPLE TO CALCULATE ENERGIES

# calculate gravitational potential energy
Model['GPE'] = m * g * Model['y']

# calculate kinetic energy
Model['KE'] = 0.5 * m * Model['vy']**2

# calculate total mechanical energy
Model['TME'] = Model['GPE'] + Model['KE']

Model.head(3)
```
