```python
import math
#from math import cos,sin,pi
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
#Interactive Input Data
nv=int(input("Number of Vectors:"))
Lang=[]
Lmag=[]
for i in range(nv):
  ELang= float(input("Enter direction vector "+str(i+1)+" in degrees: "))
  ELmag= float(input("Enter magnitud vector "+str(i+1)+": "))
  Lang.append(ELang)
  Lmag.append(ELmag)
```

```python
#Direct and Fixed Input Data
Lang=[45,120,210] #Direction
Lmag=[30,20,10]   #Magnitude
nv=min(len(Lang),len(Lmag))
#print(nv)
```

```python
print("Direction List:",Lang)
print("Magnitude List:",Lmag)

Lvy=[]
Lvx=[]

for i in range(nv):
  Ly=Lmag[i]*math.sin(math.radians(Lang[i]))
  Lx=Lmag[i]*math.cos(math.radians(Lang[i]))
  Lvx.append(Lx)
  Lvy.append(Ly)
```

```python
xdim=round(max(np.fabs(Lvx)))
ydim=round(max(np.fabs(Lvy)))
fig, ax = plt.subplots()
ax.quiver(np.zeros(nv),np.zeros(nv),Lvx,Lvy,angles='xy',scale_units='xy',scale=1,color='b')
ax.set_xlim([-xdim,xdim])
ax.set_ylim([-ydim,ydim])
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.set_title('Multiple Vectors')
ax.grid(True)
data2= pd.DataFrame({"Lang": Lang, "Lmag": Lmag, "Lvx": Lvx, "Lvy": Lvy})  #Create DataFrame table
#create right table
plt.table(cellText=data2.values.round(3), colLabels=data2.columns, loc='right')
plt.show()
```

```python

print("X Component Vector List:")
print(['%.2f' % elem for elem in Lvx])
SLvx=sum(Lvx)
print("Resultant Sum X Component Vector List: %.2f"%SLvx)
print("Y Component Vector List:")
print(['%.2f' % elem for elem in Lvy])
SLvy=sum(Lvy)
print("Resultant Sum Y Component Vector List: %.2f"%SLvy)
```

```python
LvMag=math.hypot(SLvx, SLvy)
print("Magnitud Resultant Vector: %.2f" %LvMag)
LvAng = math.degrees(math.atan2(SLvy, SLvx))
if LvAng < 0:
  print("Direction Resultant Vector: %.2f" %LvAng)
# Convert (-180 deg,180 deg] to [0 deg,360 deg)
LvAng = LvAng - 360* math.floor(LvAng/360)
#if LvAng < 0:
#  LvAng = LvAng + 360
print("Direction Resultant Vector: %.2f"%LvAng)
print("Location of the Resultant Vector Quadrant")
if LvAng >= 0 and LvAng < 90:
    print('Quadrant 1')
elif LvAng >=90 and LvAng < 180:
    print('Quadrant 2')
elif LvAng >= 180 and LvAng < 270:
    print('Quadrant 3')
else:
    print('Quadrant 4')
```

```python
LvMagO=math.hypot(SLvx, SLvy)
print("Magnitud Resultant Opposite Vector: %.2f" %LvMagO)
LvAngO = math.degrees(math.atan2(-SLvy, -SLvx))
if LvAngO < 0:
  print("Direction Resultant Opposite Vector: %.2f" %LvAngO)
# Convert (-180 deg,180 deg] to [0 deg,360 deg)
LvAngO = LvAngO - 360* math.floor(LvAngO/360)
#if LvAng < 0:
#  LvAng = LvAng + 360
print("Direction Resultant Opposite Vector: %.2f"%LvAngO)
print("Location of the Resultant Opposite Vector Quadrant")
if LvAngO >= 0 and LvAngO < 90:
    print('Quadrant 1')
elif LvAngO >=90 and LvAngO < 180:
    print('Quadrant 2')
elif LvAngO >= 180 and LvAngO < 270:
    print('Quadrant 3')
else:
    print('Quadrant 4')
```

```python
if int(len(Lang)) == nv:
  Lang.append(LvAng)
  Lmag.append(LvMag)
  Lvx.append(SLvx)
  Lvy.append(SLvy)
  Lang.append(LvAngO)
  Lmag.append(LvMagO)
  Lvx.append(-SLvx)
  Lvy.append(-SLvy)

xdim=round(max(np.fabs(Lvx)))
print(['%.2f' % elem for elem in Lvx])
#print(Lvx)
print(xdim)
ydim=round(max(np.fabs(Lvy)))
print(['%.2f' % elem for elem in Lvy])
#print(Lvy)
print(ydim)
data = pd.DataFrame({"Lang": Lang, "Lmag": Lmag, "Lvx": Lvx, "Lvy": Lvy})  #Create DataFrame table
data #print DataFrame table
```

```python
fig, ax = plt.subplots()
ax.quiver(np.zeros(nv+2),np.zeros(nv+2),Lvx,Lvy,angles='xy',scale_units='xy',scale=1,color='b')
ax.quiver(np.zeros(1),np.zeros(1),SLvx,SLvy,angles='xy',scale_units='xy',scale=1, color='r')
ax.quiver(np.zeros(1),np.zeros(1),-SLvx,-SLvy,angles='xy',scale_units='xy',scale=1, color='c')
ax.set_xlim([-xdim,xdim])
ax.set_ylim([-ydim,ydim])
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.set_title('Resulting Vector and Opposite Vector')
ax.grid(True)
#create right table
plt.table(cellText=data.values.round(3), colLabels=data.columns, loc='right')
plt.show()
```

**Topic: Vectors in Physics**

**Learning objectives:**
- Calculate the vector resulting from two or more vectors.
- Determine the direction and meaning of the resulting vector.
- Carry out various practical activities to reinforce the concepts.

Duration: 2 classes (90 minutes each)

**Activities:**

1. Introduction (10 minutes)
    - Presentation of the topic and the objectives of the lesson.
    - Discussion on the importance of vectors in physics.

2. Review of previous concepts (15 minutes)
    - Quick review of the basic concepts of vectors and their components.

3. Calculation of the resulting vector (30 minutes)
    - Theoretical explanation on how to add vectors using the graphic or analytical method.
    - Practical examples for students to practice calculating the resulting vector.

4. Determination of direction and sense (20 minutes)
    - Explanation on how to determine the direction and direction of the resulting vector.
    - Practical exercises for students to apply these concepts.

5. Practical activities (30 minutes)
    - Divide the students into groups and assign them problems related to the calculation of the resultant vector.
    - The groups must solve the problems and present their solutions to the rest of the class.

6. Evaluation (15 minutes)
    - Carrying out an individual assessment to check understanding of the concepts.
    - Feedback and clarification of doubts.

7. Closing (10 minutes)
    - Summary of the key concepts learned during the lesson.
    - Motivation to continue exploring the subject of vectors in physics.
