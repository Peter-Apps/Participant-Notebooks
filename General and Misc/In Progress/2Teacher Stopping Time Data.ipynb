##**Part 4**b
Run the next block of code to calculate the slope of the trendline. How does it compare with your calculated slope? What may have caused differences? Do you need to make adjustments to your calculation?

```
#Compute the actual slope of the trendline
def best_fit_slope(x,y):
    m = (((mean(x)*mean(y)) - mean(x*y)) /
         ((mean(x)**2) - mean(x**2)))
    return m

m = best_fit_slope(x,y)
print(m,'feet/mph')
```

How many digits should your slope have? Use the code below to round your slope.

```
#round to some number of digits
print(round(m,2),'feet/mph')
```
