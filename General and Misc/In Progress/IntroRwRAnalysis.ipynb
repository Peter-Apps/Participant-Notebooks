# Imports

The "referred to" names above are short-hand references used within the program to point Python to these packages when needed.

# Data and Histograms
The below chunk of code creates a list of values and then creates a histogram, showing how frequently each value in the list appears.

```
# Create a list of values called "numbers"
numbers = [6, 4, 3, 2, 2, 3, 1, 5, 4, 2, 6, 4, 1, 1, 2, 4, 3, 2, 2, 4, 5, 6, 2, 1, 2, 2, 3, 4, 2, 2, 2, 5, 6, 4, 1, 3, 4]
# pyplot includes a function that will automatically generate a histogram of the values in the list of numbers
plt.hist(numbers, bins = 6, edgecolor = "black")
plt.title("Histogram of Numbers in the List") # Titles can be added as well
plt.xlabel("Numbers in List")
plt.ylabel("Frequency of Number")
```

# Data from Outside Sources
Manually entering data doesn't always inspire one to greatness, but it is possible to bring in outside data sources.  These often exist in a "csv" format.

Below, you'll see how to import from a web based source.  Github, in this case...

```
#Import a data file.  Files can be found here:  https://github.com/ThePAEngineer/CosmicRayData
# From that page, you can find the web address for a specific csv file listed.  Do this by clicking on the file.
# Then find the "download" button, but "right click" on it...choosing to "copy the link location"
# Paste the link within the parenthesis below...

data = pd.read_csv('https://github.com/ThePAEngineer/CosmicRayData/raw/master/30-3%20distance%201%20to%204.txt')

# Fun fact to know:  This loads the csv file as "data" and is a "pandas data frame".  This organizes the data into columns, much as you would see in a spreadsheet.
# We can inspect the first few rows of data by using the "head" function...Here looking at 3 rows...

data.head(3)
```

```
# Pandas allows you to do some fun things with this data, math wise, such as...
# Multiply each value in a particular column by 2, and then store these values within a new column
data["ch4_1By2"] = data["Time Difference ch4-ch1"]*2
# Check the result with a call to the "head" function.  The new column should appear on the far right...
data.head(3)
```

```
# You could also do some other math and statistical functions too with a column from the data, such as...
avg = data["ch4_1By2"].mean()
avg
```

```
# Or multiple values can be generated to describe the data as well
data["ch4_1By2"].describe()
```

```
# You could also focus on a particular range of values within the column too...
data_focus = data.query("ch4_1By2 > 1 & ch4_1By2 < 13")
# Running a describe shows the standard deviation value to be much narrower than the original, full data column
data_focus["ch4_1By2"].describe()
```
