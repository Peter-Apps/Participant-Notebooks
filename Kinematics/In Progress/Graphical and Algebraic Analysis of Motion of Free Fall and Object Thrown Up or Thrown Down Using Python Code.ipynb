# Sample Code Part I
This part of the code imports the software packages, sets the coefficients, and plots a scatter plot of position vs. time.
Try changing the title of the graph, the labels of the axes, colors of the points and changing the coefficients such as initial position and initial velocity.



```python

# imports software packages
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# this sets the coefficients
xi = 100
vi = 30
a = -9.8
print("initial position =", xi, "[m]")
print("initial velocity =", vi, "[m/s]")
print("acceleration =", a, "[m/s**2]")
```

```python
# makes an empty position/time data set
time_list = []
position_list = []
t = 0               # sets a starting value for time

while (t < 8):                  # runs the 4 lines below until time is not < 8
    time_list.append(t)          # saves the time value
    xf = xi + vi*t + .5*a*t**2   # calculates xf
    position_list.append(xf)     # saves xf as the position value
    t = t + 1

#   this creates a scatterplot: plt.scatter(x-variable, y-variable)
plt.scatter(time_list, position_list)
plt.title("Position vs Time for an Object Thrown up and Falling Down")
plt.xlabel("Time (sec)")
plt.ylabel("Vertical Position(m)");

```

# Sample Code Part II
This part of the code performs the curve-fit, plots the scatter plot of position vs. time and the best fitting curve as well as prints the coefficients.
Try changing the values in t = np.linspace(0, 8, num = 40) . What happens? What do these values represent?

```python

import numpy as np

# curve-fit() function imported from scipy
from scipy.optimize import curve_fit

from matplotlib import pyplot as plt

# numpy.linspace with the given arguments
# produce an array of 40 numbers between 0
# and 10, both inclusive
t = np.linspace(0, 8, num = 40)


# y is another array which stores
# The random.normal() draws random sample
# from normal (Gaussian) distribution to make
# them scatter across the base line
y = 100 + 30*t + .5*(-9.8)*(t**2) + np.random.normal(size =40 )

# Test function with coefficients as parameters
def test(t, a, b):
    return a + b*t + .5*(-9.8)*t**2

# curve_fit() function takes the test-function
# x-data and y-data as argument and returns
# the coefficients a and b in param and
# the estimated covariance of param in param_cov

param, param_cov = curve_fit(test, t, y)


print("Coefficients:")
print(param)
print("Covariance of coefficients:")
print(param_cov)

# ans stores the new y-data according to
# ans stores the new y-data according to
# the coefficients given by curve-fit() function
ans =  param[0] + param[1]*t + .5*(-9.8)*t**2

plt.scatter(time_list, position_list)
plt.title("Position vs Time for an Object Thrown up and Falling Down")
plt.xlabel("Time (sec)")
plt.ylabel("Vertical Position(m)");
plt.plot(t, y, 'o', color ='red', label ="data")
plt.plot(t, ans, '--', color ='blue', label ="curve-fit data")
plt.legend()
plt.show()
```
