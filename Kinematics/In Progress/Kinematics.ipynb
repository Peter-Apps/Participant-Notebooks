In this notebook, you will be introducted to graphing the motion of a ball in both one- and two-dimensions using Python. Don't worry about the Python for now. This will be a tool for you to use and learn as you go.

```python
# This cell only needs to be run once, but re-exectuing it doesn't hurt anything either
# imports software packages. These are libraries that have useful functions that allow us to perform calculations and create graphs and charts.
# Notice that text with a # in front is a comment that the computer ignores. Comments are very important when sharing code with other humans.
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
plt.style.use('classic')
```

```python
# This sets the coefficients
""" The three quotes allow you to make multiline comments.
xi = 1
vi = 0
a = 3
"""

xi = float(input("What is the initial position in meters? "))  # Asks the user to input the initial position. Float allows the user to input decimal values.
vi = float(input("What is the initial velocity in meters/second? "))
a = float(input("What is the accleration in meters/second\u00b2? "))
```

```python
time_list = [] # makes an empty time data set
position_list = []
velocity_list =[]
acceleration_list = []
time_var = 25 # A general value to max time.
t = 0               # sets a starting value for time

while (t < time_var):                  # runs the 4 lines below until time is not < 11
    time_list.append(t)          # saves the time value
    xf = xi + vi*t + .5*a*t**2  # calculates xf
    v_final = vi+a*t # calculates the final velocity
    position_list.append(xf)     # saves xf as the position value
    velocity_list.append(v_final) #
    t = t + 1

#   this creates a scatterplot: plt.scatter(x-variable, y-variable)
plt.scatter(time_list, position_list)
plt.title("Position vs Time")
plt.xlabel("time (s)")
plt.ylabel("Position (m)")
plt.xlim(0, time_var)
#plt.ylim(0,20)
ptdata = pd.DataFrame({"x (m/s)": position_list, "t (s)": time_list})  #Create an array from time and position lists
plt.table(cellText=ptdata.values.round(3), colLabels=ptdata.columns, colWidths = [0.25,0.25], loc='right', cellLoc='center') #print array list and round to 3
```

Now let's see what the velocity of this ball looks like over the same time.
This time, the program will print both graphs, one above the other. This lets you see the velocity at each position.

```python
plt.figure()  # create a plot figure

# create the first of two panels and set current axis
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(time_list, position_list)

# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.xlabel('Time (s)')
plt.ylabel("Position (m)")
plt.ylabel("Velocity (m/s)")
plt.plot(time_list, velocity_list);
```
