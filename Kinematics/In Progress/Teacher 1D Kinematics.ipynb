# **How did Superman and the Flash differ?**

### Grodd is currently trying to destroy the city.  When the call came in Superman was flying over exit 250 (measured in km) at 455 m/s and just continued at the same speed onto the scene of the attack while The Flash was sitting and eating lunch and then took off straight toward the scene, accelerating at 3.78 m/s^2.  Your task is to use the following information and analyze the motion of both superheros as they travel to the scene.


* The Hall of Justice is located right off of exit 32 (measured in km)
* The distruction is taking place at exit 141 (measured in km)
* Both men left at 11:00 am




```
# These are some software packages we'll use to allow us to analyze this problem (Remember to run after each code section is complete)
import pandas as pd
import numpy as np
from math import cos,sin,pi
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```
#Defining Lists and Equations used for Superman
xi = 250000    #the initial position of Superman
xf = 141000   #the final position of Superman
vs = -455      #the velocity of Superman
sa = 0        #the acceleration of Superman
tdelta = 10   #the periods of time we will calculate data
t = 0         #the initial time
S = xi    #the position of Superman at any moment in time
Sv = vs   #the velocity of Superman at any moment in time
tlist = [] #create a list for the time values
Slist = [] #create a list for the position values
Svlist = [] #create a list for the velocity values
tlist.append(t)
Slist.append(S)
Svlist.append(Sv)



while S >= 141000: #stop the calculations after Superman reaches Grodd
    S = .5*sa*tdelta*tdelta + vs*tdelta + S #calculating the position at any moment in time
    Slist.append(S)

    Sv = sa*tdelta + Sv #calculating the velocity at any moment in time
    Svlist.append(Sv)

    t = t + tdelta #calculating the time at each interval
    tlist.append(t)

Sdata = pd.DataFrame({"t": tlist, "S": Slist, "Sv": Svlist})
Sdata
```

```
#Defining Lists and Equations used for the Flash
Fxi = 32000
Fxf = 141000
Fvi = 0
Fa = 3.78
tdelta = 30
t = 0
Fv = Fvi
F = Fxi    #the position of the Flash
tlist = []
Flist = []
Fvlist = []
tlist.append(t)
Flist.append(F)
Fvlist.append(Fv)


while F <= 141000:

    F = .5*Fa*tdelta*tdelta + Fv*tdelta + F
    Flist.append(F)

    Fv = Fa*tdelta + Fv
    Fvlist.append (Fv)
    t = t + tdelta
    tlist.append(t)

Fdata = pd.DataFrame({"t": tlist, "F": Flist, "Fv": Fvlist})
Fdata
```
