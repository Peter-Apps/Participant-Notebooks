# Produce the plots...

```
fig, ax = plt.subplots(figsize=(12,7))
# makes the scatterplot
x = data['time']
y = data['x']
ax.scatter(x,y,s=10, zorder=10, color='blue', label="Pos")
x = data['time']
y = data['vx']
ax.scatter(x,y,s=10, zorder=10, color='red', label="Vel")
ax.set_title('Position and Velocity')
ax.set_xlabel('pos (m)/vel (m/s)')
ax.set_ylabel('pos (m)/vel (m/s)')
#ax.text(5, 125, r'Position blue, Velocity red')
ax.legend()
ax.grid(True)
plt.show()
```
