### **<font color='Green'>Recap:</font>**

To make a comparison on the motion of an object using a position-time graph we can superimpose all graphs and label them.

```python
# To create a position-time graph.

time = [0, 1, 2, 3, 4] # time values
position1 = [0, 5, 10, 15, 20] # position values
position2 = [20, 15, 10, 5, 0] # position velues in reverse order
position3 = [8, 8, 8, 8, 8] # position values; no change
position4 = [0, 1, 3, 9, 18] # position values in random increasing order

plt.plot(time, position1, label='constant velocity(+)') # plot the grap
plt.plot(time, position2, label='constant velocity(-)') # plot the grap
plt.plot(time, position3, label='At rest') # plot the grap
plt.plot(time, position4, label='Accelerating') # plot the grap
plt.xlabel('Label me with unit') # set x-axis label
plt.ylabel('Label me with unit') # set y-axis label
plt.title('Give me a title') # set title of the graph
plt.legend()
plt.show() # display the graph
```
