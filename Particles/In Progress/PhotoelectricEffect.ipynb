## Photoelectric effect and finding Planck's Constant
In 1921, Einstein was give the Nobel Prize "for his services to Theoretical Physics, and especially for his discovery of the law of the photoelectric effect." His law predicts that when light of frequency $freq$ shines on a metal electrons will be ejected from the metal with a maximum kinetic energy $$K_{max} = h f - \phi,$$
where $h$ is Planck's constant and
 $\phi$ is the <i>work function</i> of the metal.

 If we can measure the maximum kinetic energy of the electrons for light of several frequencies we should be able to determine Planck's constant and the work function by doing a linear fit.

The slope of the line will give Planck's constant and the intercept will give the work function.

````python

# Create a graph showing the data.
x = data ['freq']
y = data ['Kmax']
plt.scatter(x,y, s=35, marker='o' ,color='red' , edgecolors=None)

plot(x,y,"ro", label='Data')
plot(x,y, label='Fit')

# Display a legend, label the x and y axes and title the graph.

plt.xlabel('Frequency (Hz)')
plt.ylabel('Kinetic Energy (J)')
plt.title('Frequency versus maximum electron energy')
plot(x,y )
# Show the graph in a new window on the users screen.
legend()
plt.grid(True)
plt.show()
````

````python

def linearFunc(x,intercept,slope):
    y = intercept + slope * x
    return y

a_fit, cov=curve_fit(linearFunc,x,y)

# Assign the best-fit parameters by the curve_fit() to the variable inter and slope
inter = a_fit[0]
slope = a_fit[1]

# Display the best fit values for the slope and intercept

print(f'The slope = {slope}')
print(f'The intercept = {inter}')
````
