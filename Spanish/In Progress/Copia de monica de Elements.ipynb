```python
# makes a line plot
# plt.plot(x-variable, y-variable)
plt.plot(data['AtomicNumber'], data['AtomicRadius'])
plt.title("Atomic radius plotted against Atomic number")
plt.xlabel("Atomic Number(Z)")
plt.ylabel("Atomics Radius");#you will see that the atomic radius decreases fron left to right
```

```python
# displays the first7 rows of the data set
data.head(7)
```

```python
# makes a line plot
# plt.plot(x-variable, y-variable)
plt.plot(data['AtomicNumber'], data['Electronegativity'])
plt.title("Atomic radius plotted against Electronegativity")
plt.xlabel("Atomic Number(Z)")
plt.ylabel("Electronegativity");#at increases the atomic number (Z) is going to decrease the electronegativity due to mayor distancer between electrons and the nucleus
```

```python
# displays the first 7 rows of the data set
data.head(7)
```

```python
# makes a line plot
# plt.plot(x-variable, y-variable)
plt.plot(data['AtomicNumber'], data['Electron Affinity (kJ/mol)'])
plt.title("Atomic radius plotted against Electron Affinity")
plt.xlabel("Atomic Number(Z)")
plt.ylabel("Electron Affinity (kJ/mol)");
#you will notice that electron affinity increases upward (means atomic number decrases) for the groups and for the left to right across periods,
#because electron added to energy levels become closer to the nucleus
```

```python
# makes a line plot
# plt.plot(x-variable, y-variable)
plt.plot(data['AtomicNumber'], data['AtomicMass'])
plt.title("Atomic Mass plotted against Atomic number")
plt.xlabel("Atomic Number(Z)")
plt.ylabel("Atomics Mass") #atomic mass increases from left to right
```

```python
fig, ax = plt.subplots(1, 4, figsize=(25,3))
fig.subplots_adjust(wspace=1)

ax[0].plot(data['AtomicNumber'], data['AtomicRadius'])
ax[0].set_title("Atomic radius plotted against Atomic number")
ax[0].set(xlabel="Atomic Number(Z)", ylabel="Atomics Radius")

ax[1].plot(data['AtomicNumber'], data['Electronegativity'],color='green')
ax[1].set_title("Atomic number plotted against Electronegativity")
ax[1].set(xlabel="Atomic Number(Z)", ylabel="Electronegativity")

ax[2].plot(data['AtomicNumber'], data['Electron Affinity (kJ/mol)'], color='pink')
ax[2].set_title("Atomic number plotted against Electron Affinity (kJ/mol)")
ax[2].set(xlabel="Atomic Number(Z)", ylabel="Electron Affinity (kJ/mol)")

ax[3].plot(data['AtomicNumber'], data['AtomicMass'], color='purple')
ax[3].set_title("Atomic Mass plotted against Atomic Number")
ax[3].set(xlabel="Atomic Number(Z)", ylabel="Atomis Mass)")

plt.show()
```
