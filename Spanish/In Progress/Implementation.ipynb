#Conversión de números enteros a binarios y vice-versa

<hr/>

##Definiciones

- Los **números enteros** son aquellos números positivos y negativos, incluido el cero, que no tienen parte decimal dentro de su estructura.

- Los **números binarios** son un sistema de número base 2. La Base 2 significa que solo hay dos dígitos, 1 y 0, que corresponden a los estados de encendido y apagado que tu computadora puede entender. Probablemente estés familiarizado con la base 10: el sistema decimal.

<hr/>

##¿Cómo se convierten?

\\

- Los **números enteros** se pueden convertir a *números binarios* mediante la división sucesiva por 2.

**Ejemplo: 17 = 10001**

![Alt text](https://media.geeksforgeeks.org/wp-content/uploads/20220102111215/decimal2binary.png)

Imagen recuperdada de: https://www.geeksforgeeks.org/decimal-to-binary/

\\

- Los **números binarios** se pueden convertir en *números enteros* al enumerar los dígitos de derecha a izquierda comenzando desde cero, a cada número se le asigna la correspondiente potencia base 2 y al final se suman las potencias.

**Ejemplo: 101010 = 42**

![Alt text](https://media.geeksforgeeks.org/wp-content/uploads/20211220193034/binary2.PNG)

Imagen recuperada de: https://www.geeksforgeeks.org/binary-to-decimal/

## Parte 1: Enteros a Binarios

Escribe un programa que convierta un número entero en binario y después de que el usuario lo coloque muestre por pantalla la cadena "Tu número binario es: ", donde <num_entero> es el número binario que el usuario haya colocado convertido a entero.


```python
def enteroABinario(num):                                                        # Función para convertir números enteros a binarios

    if num >= 1:                                                                # Condicional if
        enteroABinario(num // 2)
    print(num % 2, end = '')

if __name__ == '__main__':                                                      # Código control

    print("Introduce un número entero: ")
    num_entero =   + int(input(""))                                             # Introducir número entero


    print("\nTu número en binario es: ")
    enteroABinario(num_entero)                                                  # Llamar la función

                                                                                # El código fue utilizado de la página:
                                                                                # https://www.geeksforgeeks.org/python-program-to-covert-decimal-to-binary-number/
```

## Parte 2: Binarios a Enteros

Escribe un programa que convierta un número binario en entero y después de que el usuario lo coloque muestre por pantalla la cadena "Tu número entero es: ", donde <num_binario> es el número entero que el usuario haya colocado convertido a binario.

```python
def binarioAEntero(binary):                                                     # Función para convertir números binarios a enteros

    decimal, i = 0, 0

    while(binary != 0):                                                         # Condición while
        dec = binary % 10
        decimal = decimal + dec * pow(2, i)
        binary = binary//10
        i += 1
    print(decimal)

if __name__ == '__main__':                                                      # Código control

    print("Introduce un número binario: ")
    num_binario =   + int(input(""))                                            # Introducir número binario

    print("\nTu número entero es: ")
    binarioAEntero(num_binario)                                                 # Llamar la función

                                                                                # El código fue utilizado de la página:
                                                                                # https://www.geeksforgeeks.org/binary-decimal-vice-versa-python/
```

## Explora

A continuación se muestra un programa que convierte <u> tu nombre </u> en **binario**.

```python
print("Introduce tu nombre: ")
test_str = (input(""))                                                          # Introducir string

res = ''.join(format(ord(i), '08b') for i in test_str)                          # Función para convertir tu nombre a string

print("\nTu nombre en binario es: " + str(res))                                   # Desplegar resultados

                                                                                # El código fue utilizado de la página:
                                                                                # https://www.geeksforgeeks.org/python-convert-string-to-binary/
```
